# Supervise Learning Regression


**Supervised learning regression** is a type of machine learning task where the goal is to build a model that can predict a continuous output variable based on input features. In supervised learning, the training data consists of pairs of input features and corresponding target values. The objective is to learn a mapping between the input features and the target values, so that the model can make accurate predictions on unseen data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Function to implement Linear Regression
def linear_regression(x_train, y_train):
    # Add bias term to the features
    x_train = np.insert(x_train, 0, 1, axis=1)
    
    # Calculate weights using closed-form solution (normal equation)
    x_transpose = np.transpose(x_train)
    x_transpose_dot_x = np.dot(x_transpose, x_train)
    inverse = np.linalg.inv(x_transpose_dot_x)
    x_transpose_dot_y = np.dot(x_transpose, y_train)
    weights = np.dot(inverse, x_transpose_dot_y)
    
    return weights
```

```python
# Generate random data for demonstration
np.random.seed(50)
x_train = np.random.rand(100, 1)
y_train = 2 + 3 * x_train + np.random.randn(100, 1)
```

```python
# Apply Linear Regression
weights = linear_regression(x_train, y_train)

# Print the learned weights
print("Learned weights: ", weights)

```

```python
# Visualize the data and the learned line
plt.scatter(x_train, y_train, c='b', label='Actual data')
plt.plot(x_train, weights[0] + weights[1] * x_train, c='r', label='Fitted line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
